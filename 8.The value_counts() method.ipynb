1.Value Counts

```python
import pandas as pd
```

```python
df = pd.read_csv(r"youtube_4k_channels_cleaned.csv")
```

```python
df
```

```python
# counting gender elements
# len function
len(df["Country"])
```

```python
# .count() method
df["Country"].count()
```

```python
# counting country elements by category
df["Country"].value_counts()
```

```python
# return the relative frequency (divide all values by the sum of values)
df["Country"].value_counts(normalize = True).round(2)
```
