# 1. Adding a new column

# 1.1 assign()

   when to use assign?
      

*   Add multiple coulmns in a single line of code
*   When you need to overwrite the values of an existing coulmns(best practice).

It returns a new object(a copy) with all the original columns in addition to the new ones




```python
import numpy as np
```

```python
# create random numbers for our new score columns
score1 = np.random.randint(1,101,size = 3964)
score2 = np.random.randint(1,101,size=3964)
```

```python
# create a series using the random numbers
series1 = pd.Series(score1,index=np.arange(0,3964))
series2 = pd.Series(score2,index=np.arange(0,3964))
```

```python
# using assign() to add multiple columns
Df.assign(score1=series1,score2=series2)
```

```python
Df
```

```python
# here the values are not updated basically it creates out a new object and provides the new object hence we need to overwrite the existing one
df = Df.assign(score1=series1,score2=series2)
df
```
