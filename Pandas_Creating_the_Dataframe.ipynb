# 1 Craeting a dataframe from an array

# 1.1 Option 1

```python
#Creating an array
data = np.array([[1,2],[2,3],[4,5]])
data
```

```python
#Creating a data frame
pd.DataFrame(data)
```

```python
#renaming the index and column
DataFrame=pd.DataFrame(data,index=["Row1","Row2","Row3"],columns=["Col1","Col2"])
```

```python
#visualizing the data frame
DataFrame
```

2.Creating a DataFrame from a dictionary

```python
# lists used for this example
states = ["California","Texas","Florida","New York"]
population = [3892386,578269,90728822,1878799]
```

```python
#storing lists within a dictionary
dict_states = {"states":states,"population":population}
```

```python
#Creating the dataframe
df_population = pd.DataFrame(dict_states)
```

```python
#showing the dataframe
df_population
```
